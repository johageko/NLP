```python
import os
import glob
import pytesseract
import cv2
```

### Collecting All Images in a given folder

Here, All image files in a given folder are to be recognized, and the resulting text is to be saved as a text (*.txt) file.  

```python
img_dir = "C:\\Users\\johag\\Documents\\TroisAmis\\Ch_1" # Enter Directory of all images  
data_path = os.path.join(img_dir,'*g') 
files = glob.glob(data_path) 
data = [] 
for f1 in files: 
    img = cv2.imread(f1) 
    data.append(img) 
```

```python
len(data)
```

### OCR for all of these files!

```python
#This is the folder to which all of the text recognized in the OCR process will need to be appended.
kapitel = ""
```

A for-loop over data should now allow us to extract all of the text and append it to "kapitel".

```python
for i in data: 
    my_text = pytesseract.image_to_string(i, lang='fra')
    kapitel = kapitel + my_text
```

```python
kapitel
```

### Printing to text 

```python
with open("Ch_1again.txt", "w") as text_file:
    text_file.write(kapitel)
```

In effect, the entire text of the files recognized could be recovered. Due to naming conventions, some where not placed correctly, but this was a problem of the original file names, not of the processesst in motion here.
